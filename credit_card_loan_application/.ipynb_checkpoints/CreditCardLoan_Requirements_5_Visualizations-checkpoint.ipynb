```python
from pyspark.sql import SparkSession
import os
import sys
#from pyspark import SparkContext
from pyspark.sql.types import StructType,StructField,StringType,IntegerType,DateType,DecimalType,DoubleType,BooleanType
#from pyspark.shell import spark


from pyspark.sql.functions import initcap, lower,col
#from pyspark.sql import functions as sf
from pyspark.sql.functions import regexp_replace
#import org.apache.spark.sql.functions.format_number
import json
import requests
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


os.environ['PYSPARK_PYTHON'] = sys.executable
os.environ['PYSPARK_DRIVER_PYTHON'] = sys.executable
spark = SparkSession.builder.appName('CreditCardApp').getOrCreate()
```

```python
response = requests.get("https://raw.githubusercontent.com/platformps/LoanDataset/main/loan_data.json")
loan_data = json.loads(response.text)

sp_df = spark.createDataFrame(loan_data)
```

```python
df = pd.DataFrame(loan_data)

newdf = pd.DataFrame(df,columns = ['Application_Status','Gender','Married','Income'])
newdf = newdf.query('Application_Status=="Y" and Married =="Yes"')
Application_Status = newdf['Application_Status'].value_counts()['Y']
```

Functional Requirements 5.1
Create a bar chart that shows the difference in application approvals for Married Men vs Married Women based on income ranges. (number of approvals)

```python
newdf.groupby(['Income','Gender']).size().plot(kind = "bar",figsize=(7,5),stacked=True)
plt.title('Loan Approvals for Married Men vs Women')
plt.xlabel('Income Levels and Gender')
plt.ylabel('Number of Approved applications')
```
